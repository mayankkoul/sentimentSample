```python
import pandas as pd

# Correct URL pointing to the raw version of the data
url = 'https://raw.githubusercontent.com/mayankkoul/sentimentSample/df482745cc76cad890696ecafa01b4d83af5c420/data.csv'
df = pd.read_csv(url)
```

```python
# prompt: drop the column Unnamed: 0 inplace

df.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
# prompt: Using dataframe df: make a plot of number of tweets per day you can use tweet_id for counting

import pandas as pd
import matplotlib.pyplot as plt

# Convert tweet_datetime to datetime object
df['tweet_datetime'] = pd.to_datetime(df['tweet_datetime'])

# Group by date and count the number of tweets
df_grouped = df.groupby(df['tweet_datetime'].dt.date)['id_tweet'].count()

# Plot the number of tweets per day
df_grouped.plot(kind='line')
plt.xlabel('Date')
plt.ylabel('Number of Tweets')
plt.title('Number of Tweets per Day')
plt.show()
```

```python
import nltk
nltk.download('punkt')
nltk.download('wordnet')
```

```python
import pandas as pd
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.tokenize import word_tokenize


# Initialize the stemmer and lemmatizer
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

# Function to apply stemming and lemmatization to text
def preprocess_text(text):
    # Tokenize text
    tokens = word_tokenize(text)

    # Apply stemming and lemmatization
    stemmed_tokens = [stemmer.stem(token) for token in tokens]
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]

    # Join tokens back to string
    return text, " ".join(stemmed_tokens), " ".join(lemmatized_tokens)


# Apply preprocessing to each tweet in the DataFrame
df['processed_data'] = df['tweet_content'].apply(lambda x: preprocess_text(x))
```

```python
# Example: Analyzing sentiment of lemmatized text
from textblob import TextBlob

# Assuming the DataFrame column 'processed_data' has tuples (original, stemmed, lemmatized)
tweets_with_sentiment = []

for row in df.itertuples(index=False):
    original, stemmed, lemmatized = row.processed_data  # Adjust based on actual column name
    analysis = TextBlob(lemmatized)
    tweets_with_sentiment.append((original, stemmed,lemmatized, analysis.sentiment.polarity, analysis.sentiment.subjectivity))

# Optionally, you can store the results back into a DataFrame
tweets_sentiment_df = pd.DataFrame(tweets_with_sentiment, columns=["Original", "Stemmed","Lemmatized", "Polarity", "Subjectivity"])
```

```python
# prompt: create a word cloud using the tweets sentiment df using Original column

import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Create a word cloud using the original text from the tweets
text = " ".join(df["tweet_content"].tolist())

wordcloud = WordCloud(width=800, height=600, background_color="white").generate(text)

# Display the word cloud
plt.figure(figsize=(10, 7))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
# prompt: a donut chart to show share of polarity from tweets_sentiment_df, to show how much percentage of tweets are positive, vs neutral vs  negative

import matplotlib.pyplot as plt
# Calculate the percentage of positive, neutral, and negative tweets
positive_tweets = tweets_sentiment_df[tweets_sentiment_df['Polarity'] > 0].shape[0]
neutral_tweets = tweets_sentiment_df[tweets_sentiment_df['Polarity'] == 0].shape[0]
negative_tweets = tweets_sentiment_df[tweets_sentiment_df['Polarity'] < 0].shape[0]

total_tweets = positive_tweets + neutral_tweets + negative_tweets

positive_percentage = positive_tweets / total_tweets * 100
neutral_percentage = neutral_tweets / total_tweets * 100
negative_percentage = negative_tweets / total_tweets * 100

# Create a donut chart
labels = ['Positive', 'Neutral', 'Negative']
sizes = [positive_percentage, neutral_percentage, negative_percentage]
colors = ['green', 'yellow', 'red']

fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors)
ax.axis('equal')  # Equal aspect ratio ensures a circular pie chart

# Add a title and show the chart
ax.set_title('Share of Polarity from Tweets')
plt.show()
```

```python
# prompt: a donut chart to show share of subjectivity from tweets_sentiment_df, to show how much percentage of tweets are opinion, vs neutral vs  fact

import matplotlib.pyplot as plt
# Calculate the percentage of subjective, neutral, and objective tweets
subjective_tweets = tweets_sentiment_df[tweets_sentiment_df['Subjectivity'] > 0.5].shape[0]
neutral_tweets = tweets_sentiment_df[tweets_sentiment_df['Subjectivity'] == 0.5].shape[0]
objective_tweets = tweets_sentiment_df[tweets_sentiment_df['Subjectivity'] < 0.5].shape[0]

total_tweets = subjective_tweets + neutral_tweets + objective_tweets

subjective_percentage = subjective_tweets / total_tweets * 100
neutral_percentage = neutral_tweets / total_tweets * 100
objective_percentage = objective_tweets / total_tweets * 100

# Create a donut chart
labels = ['Subjective', 'Neutral', 'Objective']
sizes = [subjective_percentage, neutral_percentage, objective_percentage]
colors = ['green', 'yellow', 'red']

fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors)
ax.axis('equal')  # Equal aspect ratio ensures a circular pie chart

# Add a title and show the chart
ax.set_title('Share of Subjectivity from Tweets')
plt.show()
```
